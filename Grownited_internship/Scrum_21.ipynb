```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data provided in the problem description
data = {
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [30000, 45000, 55000, 70000, 50000]
}
df = pd.DataFrame(data)

print("Employee Salary Data:")
display(df)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Salary'], kde=True, bins=5, color='skyblue')
plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Salary'], color='lightgreen')
plt.title('Box Plot of Employee Salaries')
plt.ylabel('Salary')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Employee', y='Salary', data=df, palette='viridis')
plt.title('Employee Salaries')
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
