```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data
data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Priya', 'Jay'],
    'Marks': [70, 92, 88, 65, 45]
}
df = pd.DataFrame(data)

print("Student Data:")
display(df)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='Marks', data=df, palette='viridis')
plt.title('Student Marks - Bar Chart')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Marks'], labels=df['Name'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Student Marks Distribution - Pie Chart')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Name', y='Marks', data=df, marker='o', color='skyblue')
plt.title('Student Marks - Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
