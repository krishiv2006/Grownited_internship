```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create ipl_players.csv
players_data = """Player,Team,Matches,Runs,Wickets,Sixes,Fours,StrikeRate
Virat Kohli,RCB,14,741,0,38,62,154.69
Rohit Sharma,MI,14,520,0,28,44,145.32
Shubman Gill,GT,14,610,0,24,58,151.20
KL Rahul,LSG,14,480,0,20,40,136.50
Ruturaj Gaikwad,CSK,14,580,0,25,52,148.75
Andre Russell,KKR,13,350,18,30,22,185.50
Jasprit Bumrah,MI,14,25,22,0,2,105.00
Mohammed Shami,GT,14,15,24,0,1,90.00
Sunil Narine,KKR,14,320,16,22,25,178.40
Ravindra Jadeja,CSK,14,270,15,12,18,140.20
"""

with open('ipl_players.csv', 'w') as f:
    f.write(players_data)

# Create ipl_matches.csv
matches_data = """Team,Matches,Won,Lost,Points
KKR,14,9,5,18
SRH,14,8,6,16
RR,14,8,6,16
RCB,14,7,7,14
CSK,14,7,7,14
DC,14,7,7,14
LSG,14,7,7,14
GT,14,5,9,10
PBKS,14,5,9,10
MI,14,4,10,8
"""

with open('ipl_matches.csv', 'w') as f:
    f.write(matches_data)

print("CSV files 'ipl_players.csv' and 'ipl_matches.csv' created successfully.")

# Load the datasets
df_players = pd.read_csv('ipl_players.csv')
df_matches = pd.read_csv('ipl_matches.csv')

print("\nFirst 5 rows of ipl_players.csv:")
display(df_players.head())

print("\nFirst 5 rows of ipl_matches.csv:")
display(df_matches.head())
```

### Task 1: Display Top 5 Run Scorers

```python
top_run_scorers = df_players.sort_values(by='Runs', ascending=False).head(5)
print("Top 5 Run Scorers:")
display(top_run_scorers[['Player', 'Runs']])
```

### Task 2: Display Top 5 Wicket Takers

```python
top_wicket_takers = df_players.sort_values(by='Wickets', ascending=False).head(5)
print("Top 5 Wicket Takers:")
display(top_wicket_takers[['Player', 'Wickets']])
```

### Task 3: Display Team Points Table

```python
team_points_table = df_matches.sort_values(by='Points', ascending=False)
print("Team Points Table:")
for index, row in team_points_table.iterrows():
    print(f"{row['Team']} {row['Points']}")
```

### Task 4: Find the Player with Highest Strike Rate

```python
highest_sr_player = df_players.loc[df_players['StrikeRate'].idxmax()]
print("Player with Highest Strike Rate:")
print(f"{highest_sr_player['Player']}\nStrike Rate : {highest_sr_player['StrikeRate']:.2f}")
```

### Task 5: Find the Team with Highest Points

```python
highest_points_team = df_matches.loc[df_matches['Points'].idxmax()]
print("Team with Highest Points:")
print(f"{highest_points_team['Team']}\nPoints : {highest_points_team['Points']}")
```

### Task 6: Calculate Average Runs

```python
average_runs = df_players['Runs'].mean()
print(f"Average Runs : {average_runs:.1f}")
```

### Task 7: Find Players Having More Than 500 Runs

```python
players_over_500_runs = df_players[df_players['Runs'] > 500]
print("Players Having More Than 500 Runs:")
for player in players_over_500_runs['Player']:
    print(player)
```

### Task 8: Find All-Rounders

```python
all_rounders = df_players[(df_players['Runs'] > 300) & (df_players['Wickets'] > 10)]
print("All-Rounders:")
for player in all_rounders['Player']:
    print(player)
```

## Visualization Tasks

### Chart 1: Top 5 Run Scorers Bar Chart

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Runs', data=top_run_scorers, palette='viridis')
plt.title('Top 5 Run Scorers')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Chart 2: Top 5 Wicket Takers Bar Chart

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Wickets', data=top_wicket_takers, palette='magma')
plt.title('Top 5 Wicket Takers')
plt.xlabel('Player')
plt.ylabel('Wickets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Chart 3: Team Points Table Bar Chart

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Team', y='Points', data=team_points_table, palette='coolwarm')
plt.title('Team Points Table')
plt.xlabel('Team')
plt.ylabel('Points')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Chart 4: Runs Distribution Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_players['Runs'], bins=5, kde=True, color='skyblue')
plt.title('Runs Distribution')
plt.xlabel('Runs')
plt.ylabel('Number of Players')
plt.tight_layout()
plt.show()
```

### Chart 5: Strike Rate vs Runs Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runs', y='StrikeRate', data=df_players, hue='Player', size='Runs', sizes=(50, 500), palette='deep', legend='full')
plt.title('Strike Rate vs Runs')
plt.xlabel('Runs')
plt.ylabel('Strike Rate')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Chart 6: Team Winning Percentage Pie Chart

```python
df_matches['Winning_Percentage'] = (df_matches['Won'] / df_matches['Matches']) * 100

plt.figure(figsize=(10, 10))
plt.pie(df_matches['Winning_Percentage'], labels=df_matches['Team'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Team Winning Percentage')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

### Chart 7: Correlation Heatmap

```python
correlation_cols = ['Runs', 'Wickets', 'Sixes', 'Fours', 'StrikeRate']
correlation_matrix = df_players[correlation_cols].corr()

plt.figure(figsize=(8, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Player Statistics')
plt.tight_layout()
plt.show()
```

## Final Report Output

```python
total_players = len(df_players)
top_scorer_player = top_run_scorers.iloc[0]['Player']
top_scorer_runs = top_run_scorers.iloc[0]['Runs']
top_bowler_player = top_wicket_takers.iloc[0]['Player']
top_bowler_wickets = top_wicket_takers.iloc[0]['Wickets']
best_team = highest_points_team['Team']
best_team_points = highest_points_team['Points']
highest_sr_player_name = highest_sr_player['Player']
highest_sr_value = highest_sr_player['StrikeRate']
all_rounders_names = all_rounders['Player'].tolist()
average_runs_calculated = df_players['Runs'].mean()

report_output = f"""
=========================================
IPL ANALYTICS REPORT
=========================================
Total Players : {total_players}
Top Scorer : {top_scorer_player}
Runs : {top_scorer_runs}
Top Bowler : {top_bowler_player}
Wickets : {top_bowler_wickets}
Best Team : {best_team}
Points : {best_team_points}
Highest Strike Rate :
{highest_sr_player_name}
Strike Rate : {highest_sr_value:.2f}
All-Rounders :
"""
for player in all_rounders_names:
    report_output += f"{player}\n"
report_output += f"Average Runs :\n{average_runs_calculated:.1f}\n"
report_output += f"""
=========================================
END OF REPORT
========================================="""

print(report_output)
```
